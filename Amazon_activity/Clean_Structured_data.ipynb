# Now we check -> data is inconsistent (mixed typed), missing data (age),duplicate(alice twice hai )

```python
# clean & structured data
def clean_data(data):
    text_to_num={"one":1,"two":2,"three": 3,"four":4,"five":5}
    for user in data:
        # clean rating part
        raw_rating=str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating=text_to_num[raw_rating]
       # print(raw_rating)
        user["rating"]=raw_rating
        
```

```python
clean_data(data1)
```

```python
# handling missing value(age)
def missing_value(data):
   
    for user in data:
        
        raw_age= user.get("age")
        
        if(raw_age ==None):
            user["age"]=None
       # print(user["age"])
        
```

```python
missing_value(data1)
```

```python
# remove duplicate (alice)
def duplicate_remove(data):
    clean_data=[]
    unique_users=set()

    for user in data:
        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        clean_data.append(user)

    return clean_data
        
        
    
```

```python
duplicate_remove(data1)
```

```python
# To get mening insight from data
def get_insights(data):

    tot_rating=0
    for user in data:
        tot_rating+=float(user["rating"])

    print(f"avg rating ={tot_rating /len (data)}")

    # percentage of users with poor rating
    poor_rating=0
    for user in data:
        if(float(user["rating"]) <3):
            poor_rating+=1
    print(f" % of user with poor rating = {poor_rating /len(data) *100} %")
    
```

```python
get_insights(data1)
```

# Build a product recommendation feature(rule based not Ml)
# 1. if user rating >=4 , recommend same brand products
# 2. f user rating < 4 , recommend different brand products


```python
def get_recommendations(data):
    recommendations=[]
    

    for user in data:
        curr_recomm={}
        curr_recomm["name"]=user["name"]
        if(float(user["rating"]) >=4):
            curr_recomm["brand"]="Apple"
        else:
            curr_recomm["brand"] ="Samsung"
        recommendations.append(curr_recomm)
    return recommendations
```

```python
get_recommendations(data1)
```
